```python
coll={}
newcol={}
for j in data[1:]:
    i=j.split(',')
    coll[i[0]]=[i[9],float(i[14]),float(i[15])]
    newcol[i[9]]=(i[9] in newcol) and newcol[i[9]]+[[float(i[14]),float(i[15])]] or [[float(i[14]),float(i[15])]]
print newcol
```

```python
import json
file('data.txt','w').write(json.dumps(coll))
```

```python
class node:
    def __init__(self,x1,y1,x2,y2,d,pois=-1):
        '''left lower right upper direction'''
        self.x1,self.y1,self.x2,self.y2,self.d=x1,y1,x2,y2,d
        if pois==-1:
            self.pois=[]
            for i in coll:
                if self.inside(coll[i][1],coll[i][2]):
                    self.pois+=[coll[i]]
        else: self.pois=pois
        self.n=len(self.pois)
        if len(self.pois)<5:
            self.left,self.right=-1,-1
        else:
            if d==0:
                tmp=sorted(self.pois,key=lambda x:x[1])
                midx=tmp[self.n/2][1]+1e-3*(x2-x1)
                self.left=node(x1,y1,midx,y2,1-d,tmp[:self.n/2+1])
                self.right=node(midx,y1,x2,y2,1-d,tmp[self.n/2+1:])
            else:
                tmp=sorted(self.pois,key=lambda x:x[2])
                midy=tmp[self.n/2][1]+1e-3*(y2-y1)
                self.left=node(x1,y1,x2,midy,1-d,tmp[:self.n/2+1])
                self.right=node(x1,midy,x2,y2,1-d,tmp[self.n/2+1:])
    def __getitem__(self,ins):
        x,y,ths=ins
        dis=lambda x1,y1,x2,y2:((x1-x2)**2+(y1-y2)**2)**0.5
        if (x<self.x1 and self.x1-x>ths) or (y<self.y1 and self.y1-y>ths)\
            or (x>self.x2 and x-self.x1>ths) or (y>self.y2 and y-self.y1>ths)\
            or (not self.inside(x,y) and min([dis(x,y,self.x1,self.y1),dis(x,y,self.x1,self.y2),\
                                              dis(x,y,self.x2,self.y1),dis(x,y,self.x2,self.y2)])>ths):
            return []
        if self.left==-1:
            return [i for i in self.pois if dis(x,y,i[1],i[2])<=ths]
        return self.left[x,y,ths]+self.right[x,y,ths]
    def inside(self,x,y):
        return x>=self.x1 and x<=self.x2 and y>=self.y1 and y<=self.y2
tmp=sorted(coll.values(),key=lambda x:x[1])
x1,x2=tmp[0][1],tmp[-1][1]
tmp=sorted(coll.values(),key=lambda x:x[2])
y1,y2=tmp[0][2],tmp[-1][2]
mytree=node(x1,y1,x2,y2,0)
print x1,y1,x2,y2
```
